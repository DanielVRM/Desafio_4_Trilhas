```python
from functools import reduce
import pandas as pd
```

```python
def corrigirDados(dados, ano):
    'ESSA FUNÇÃO RETIRA O (Maranhão) do campo.'
    'Exemplo: PARNA DAS NASCENTES DO RIO PARNAÍBA (MARANHÃO) --> PARNA DAS NASCENTES DO RIO PARNAÍBA'
    for x,y in enumerate(dados.Campo):
        dados.loc[x,"Campo"] = y.replace(" (MARANHÃO)", "")
    for x,y in enumerate(dados.Campo):
        dados.loc[x,"Campo"] = y.replace("TI ", "")
    dados.rename(columns={"Valor": ano, "Porcentagem do Total de Focos": f"Porcentagem do Total de Focos {ano}"}, inplace=True)
```

## Carregando Dados

```python
dados_2011 = pd.read_csv('dados/Focos por TI - de 2011-01-01 00_00 a 2011-12-31 23_59.csv', sep = ',')
corrigirDados(dados_2011, 2011)
```

## Fazendo Merge dos dados

```python
dados_terras_indigenas_incendio_2011_2020 = reduce(lambda a, b: pd.merge(a, b, how = 'outer', on = 'Campo'), lista_dados )
```
