## Análise Queimadas

### Importando base de dados

```python
import pandas as pd
```

```python
queimadas_municipio = pd.read_excel('dados/total_focos_queimadas_por_estado.xlsx')
```

```python
queimadas_municipio.columns = ['Municipios','2000', '2001','2002','2003',
               '2004','2005','2006','2007','2008', '2009','2010','2011','2012','2013',
               '2014','2015','2016','2017','2018','2019','2020']

queimadas_municipio = queimadas_municipio.fillna(0)
```

```python
queimadas_municipio['Total_Queimadas_Anos'] = queimadas_municipio.iloc[:,1:].sum(axis=1)
queimadas_municipio
```

```python
queimadas_municipio = queimadas_municipio.sort_values(by='Total_Queimadas_Anos', ascending = False)
queimadas_municipio.index = range(218)
```

```python
municipios = ['Amarante do Maranhão ','Arame ','Barra do corda ', 'Bom Jardim ',
'Bom Jesus das Selvas ',
'Buriticupu ',
'Fernando Falcão ',
'Grajaú ',
'Jenipapo dos Vieiras ',
'Monção ',
'Santa Luzia ']

municipio_aldeia = queimadas_municipio.query('Municipios in @municipios')
```

```python
municipio_aldeia[['Municipios', 'Total_Queimadas_Anos']]
```

```python
top5_queimadas = municipio_aldeia.iloc[:5,:-1]
top5_queimadas.index.name = 'Municipios'
lista = list(top5_queimadas['Municipios'])
lista
top5_queimadas.index = lista
top5_queimadas.drop(['Municipios'], axis = 1, inplace = True)
top5_queimadas
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
```

```python
import itertools

transform = pd.DataFrame(columns=['Municipio', 'Ano', 'Numero de Focos'])
for Municipio, ano in itertools.product(top5_queimadas.index, top5_queimadas.columns):
    numero = top5_queimadas.at[Municipio, ano]
    transform = transform.append({'Municipio': Municipio, 'Ano': ano, 'Numero de Focos': numero},
                           ignore_index=True)
    
transform.head(100)
```

```python
    plt.figure(figsize=(100,1000))
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    sns.relplot(x='Ano',y='Numero de Focos',hue='Municipio', kind = 'line', data = transform)
```
