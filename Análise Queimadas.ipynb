```python
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
```

```python
import itertools

transform = pd.DataFrame(columns=['Municipio', 'Ano', 'Numero de Focos'])
for Municipio, ano in itertools.product(top5_queimadas.index, top5_queimadas.columns):
    numero = top5_queimadas.at[Municipio, ano]
    transform = transform.append({'Municipio': Municipio, 'Ano': ano, 'Numero de Focos': numero},
                           ignore_index=True)
    
transform.head(100)
transform['Ano'] = transform['Ano'].astype(int)
transform = transform.drop(transform[transform['Ano'] < 2008].index)
transform.head(30)
```

```python
def plotar(titulo, labelx, labely, x, y, hue, dataset):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x=x,y=y, hue = hue, data = dataset)
    ax.figure.set_size_inches(18,10)
    ax.set_title(titulo, loc = 'left', fontsize = 18)
    ax.set_xlabel(labelx, fontsize = 14)
    ax.set_ylabel(labely, fontsize = 14)
    ax = ax

plotar('Os 5 municípios envolvendo aldeias indígenas com mais focos de queimada(2008 - 2020)', 'Anos', 'Numero de Focos', 'Ano', 'Numero de Focos', 'Municipio', transform)
```
